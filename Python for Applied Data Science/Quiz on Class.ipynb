```python
from math import pi
class circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return pi * self.radius ** 2
    def circumference(self):
        return 2 * pi * self.radius

        
```

```python
my_circle = circle(2)
my_circle.area()
```

```python
class car:
    max_speed = 120  #km/hr
    def __init__(self, make, model, color, speed=0):
        self.make = make
        self.model = model
        self.color = color
        self.speed = speed
    def accelerate(self, acceleration):
        self.New_Speed = acceleration + self.speed
        if self.New_Speed <= self.max_speed:
            self.speed += self.New_Speed
        else:
            self.speed = self.max_speed
    def get_speed(self):
        return self.speed
```

```python
car1 = car("Toyota", "Camry", "Blue")
car2 = car("Honda", "Civic", "Red")
car1.accelerate(30)
car2.accelerate(20)


print(f"The speed of {car1.make} {car1.model} is {car1.get_speed()} km/hr")
print(f"The speed of {car2.make} {car2.model} is {car2.get_speed()} km/hr")
```

```python
import matplotlib.pyplot as plt
class rectangle:
    def __init__ (self, Length, Width, Color):
        self.Length = Length
        self.Width = Width
        self.Color = Color
    def change_size(self, new_length, new_width):
        self.Length = new_length
        self.Width = new_width
    def change_color(self, new_color):
        self.Color = new_color
    def draw_rectangle(self):
        plt.figure(figsize=(self.Length, self.Width))
        plt.gca().add_patch(plt.Rectangle((0, 0), self.Length, self.Width, color=self.Color))
        plt.xlim(0, self.Length + 1)
        plt.ylim(0, self.Width + 1)
        plt.axis()
        plt.show()

       
```

```python
my_rectangle = rectangle(4, 2, 'purple')
my_rectangle.draw_rectangle()
```

```python
class ProjectEquipment:
    def __init__(self):
        # We start with a blank slate
        self.inventory = {}

    def add_item(self, name, quantity):
        if name in self.inventory:
            # If it's already there, ADD to the existing total
            self.inventory[name] += quantity
            print(f"Updated {name} quantity: {self.inventory[name]}")
        else:
            # If it's new, create the entry
            self.inventory[name] = quantity
            print(f"Added {name} with quantity: {quantity}")

    def remove_item(self, name, quantity):
        if name in self.inventory:
            # Subtract the amount
            self.inventory[name] -= quantity
            
            # Constraint: If it goes below zero, reset to 0
            if self.inventory[name] < 0:
                self.inventory[name] = 0
        else:
            print("Cannot remove item: Not in inventory.")

    def check_stock(self, name):
        if name in self.inventory:
            return self.inventory[name]
        else:
            return "Item not found"
```

```python
Q = ProjectEquipment()
Q.add_item("Laptop", 5)
```
