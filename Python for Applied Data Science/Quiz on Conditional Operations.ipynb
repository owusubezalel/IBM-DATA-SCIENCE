Write a Python program to check if a player Lionel Messi has more than 10 achievements. If the condition is true, print the player's name, sport, and achievements else print does not have more than 10 achievements.

```python
Player_Name = ['Serena Williams', 'Lionel Messi', 'Michael Phelps', 'Usain Bolt',  'Roger Federer', 'Cristiano Ronaldo']
Sport = ['Tennis', 'Soccer', 'Swimming', 'Athletics', 'Tennis', 'Soccer']
Achievements = [23, 7, 23, 8, 20, 5]
Achievements_Type = ['Grand Slams', "Ballon d'Ors", 'Gold Medals', 'Gold Medals', 'Grand Slams', "Ballon d'Ors"]
```

```python
Your_Player = str(input("Enter the name of a player: "))
index_player = Player_Name.index(Your_Player)
His_Achievements = Achievements[index_player]
His_Achievements_Type = Achievements_Type[index_player]
His_Sport = Sport[index_player]

if His_Achievements > 10:
    print(f"{Your_Player} is into {His_Sport} and has more than 10 achievements in {His_Achievements_Type}.")
else:
    print(f"{Your_Player}'s sport is {His_Sport} and has less than 10 achievements in {His_Achievements_Type}.")

```
