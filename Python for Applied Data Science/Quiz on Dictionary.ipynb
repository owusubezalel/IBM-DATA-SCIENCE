# Scenario:Inventory Store

The inventory store scenario project utilizes a dictionary-based approach to develop a robust system for managing and tracking inventory in a retail store.
Note:- You will be working with two product details.

## Task-1 Create an empty dictionary

```python
inventory_store ={}
```

First you need to create an empty dictionary, where you will be storing the product details.

## Task-2 Store the first product details in variable
+ Product Name= Mobile phone
+ Product Quantity= 5
+ Product price= 20000
+ Product Release Year= 2020

```python
Product_Name_1 = 'Mobile Phone'
Product_Quantity_1 = 5
Product_Price_1 = 20000
Product_Release_Year_1 = 2020
```

## Task-3 Add the details in inventory

```python
inventory_store["Product_Name_1"] = Product_Name_1
inventory_store["Product_Quantity_1"] = Product_Quantity_1
inventory_store["Product_Price_1"] = Product_Price_1
inventory_store["Product_Release_Year_1"] = Product_Release_Year_1
```

## Task-4 Store the second product details in a variable.
+ Product Name= "Laptop"
+ Product Quantity= 10
+ Product price = 50000
+ Product Release Year= 2023

```python
Product_Name_2 = 'Laptop'
Product_Quantity_2 = 10
Product_Price_2 = 50000
Product_Release_Year_2 = 2023
```

## Task-5 Add the item detail into the inventory

```python
inventory_store["Product_Name_2"] = Product_Name_2
inventory_store["Product_Quantity_2"] = Product_Quantity_2
inventory_store["Product_Price_2"] = Product_Price_2
inventory_store["Product_Release_Year_2"] = Product_Release_Year_2
```

## Task-6 Display the Products present in the inventory
