```python
my_string = "Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
```

```python
class TextChecker(object):
    def __init__(self, my_string):
        self.my_string = my_string
        formatText = self.my_string.replace("!","").replace(".","").replace("?","").replace(",","").lower()
        self.my_string = formatText.split()
    def FreqWordForALL(self):
        ListOfWords = set(self.my_string)
        Dict = {}
        for word in ListOfWords:
            Dict[word] = self.my_string.count(word)
        return Dict
    def FreqWord(self, word):
        NDict = self.FreqWordForALL()
        if word in NDict:
            return NDict[word]
        else:
            return 0
        
    
```

```python
Magna = TextChecker(my_string)
Magna.FreqWord('magna')
```

```python
flow_rates = [12.5, 15.2, 12.5, 18.9, 11.4, 15.2, 25.4, 10.1, 12.5, 30.2, 11.4]
```

```python
class HydroStat():
    def __init__(self, flow_rates):
        self.flow_rates = flow_rates
    def get_average(self):
        return sum(self.flow_rates) / len(self.flow_rates)
    def extrenes(self):
        my_list = self.flow_rates
        Dictionary ={}
        for i in my_list:
            Dictionary['maximum Flow Rate'] = max(my_list)
            Dictionary['minimum Flow Rate'] = min(my_list)
            return Dictionary
    def threshold_check(self, limit):
        my_list = self.flow_rates
        my_new_list = []
        for i in my_list:
            if i > limit:
                my_new_list.append(i)
        return my_new_list
       
```

```python
Hydr = HydroStat(flow_rates)
Hydr.threshold_check(18)
```

```python
text1 = "The watershed contains a heavy clay soil base. This soil has low permeability and high runoff risk during storm events."
text2 = "Engineers found that the site has a sandy soil base. High permeability was recorded, which reduces the overall runoff risk."
```

```python
class DocumentAnalyzer():
    def __init__(self,text1,text2):
        self.text1 = text1
        self.text2 = text2
        self.FormattedText1 = self.text1.replace(".","").replace(",","").lower()
        self.FormattedText2 = self.text2.replace(".","").replace(",","").lower()
        self.set1 = set(self.FormattedText1.split())
        self.set2 = set(self.FormattedText2.split())
    def common_words(self):
        return self.set1 & self.set2
    def unique_words(self):
        return self.set1 - self.set2
```

```python
CheckText = DocumentAnalyzer(text1,text2)
CheckText.common_words()
CheckText.unique_words()
```
